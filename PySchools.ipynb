```python
import pandas as pd 
```

```python
school = "raw_data/schools_complete.csv"
students = "raw_data/students_complete.csv"
school_pd = pd.read_csv(school,encoding ='iso-8859-1')
students_pd = pd.read_csv(students, encoding = 'iso-8859-1')
math_list =[]
reading_list=[]
```

```python
school_rename = school_pd.rename(columns={"name": "School Name"})
students_rename = students_pd.rename(columns={"school": "School Name"})
```

```python
merged_tables = pd.merge(school_rename, students_rename, on="School Name", how = "outer")
```

```python
total_schools = school_pd["School ID"].count()
total_students = students_pd["name"].count()
total_budget = school_pd["budget"].sum()
Average_Math = students_pd["math_score"].mean()
Average_Reading = students_pd["reading_score"].mean()
```

```python
for x in merged_tables["math_score"]:
    if x >=70:
        counter=1
    else:
        counter=0
    math_list.append(counter)
    counter=0
merged_tables["math"]=math_list
math_percentage=(merged_tables["math"].sum()/total_students)*100
math_percentage
```

```python
for x in merged_tables["reading_score"]:
    if x>=70:
        counter=1
    else:
        counter=0
    reading_list.append(counter)
merged_tables["reading"]=reading_list
reading_percentage=(merged_tables["reading"].sum()/total_students)*100
reading_percentage
```

```python
overall_pass = (reading_percentage+math_percentage)/2
overall_pass
```

```python
district_summary_table=pd.DataFrame({"Total Students": total_students,
                                    "Total Budget": total_budget,
                                    "Average Math Score": Average_Math,
                                    "Average Reading Score": Average_Reading,
                                    "% Passing Math": math_percentage,
                                    "% Passing Reading": reading_percentage,
                                    "% Overall Passing Rate": overall_pass},index=[0])
district_summary_table
```

```python
summary_table = merged_tables.groupby(["School Name", "type"])
print(summary_table)
summary_table.count()


```

```python
students=summary_table["name"].count()
budget =summary_table["budget"].mean()
budget_per_student= budget/students

avg_math=summary_table['math_score'].mean()
avg_reading=summary_table['reading_score'].mean()
math_count=summary_table["math"].sum()
read_count=summary_table["reading"].sum()
percent_pass_math=(math_count/students)*100
percent_pass_read=(read_count/students)*100
overall_pass=(percent_pass_math+percent_pass_read)/2
```

```python
School_summary_table=pd.DataFrame({
                                    "% Overall Passing Rate": overall_pass,
                                    "% Passing Reading": percent_pass_read,
                                    "% Passing Math": percent_pass_math,
                                    "Average Reading Score": avg_reading,
                                    "Per Student Budget": budget_per_student,
                                    "Average Math Score": avg_math,
                                    "Total Budget": budget,
                                    "Total Students": students
                                  })
School_summary_table
```

```python
top_schools=School_summary_table.sort_values("% Overall Passing Rate", ascending=False)
top_schools.head(5)
```
